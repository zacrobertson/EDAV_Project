# Introduction

As technology has progressed and allowed people to consume news in ways that previous generations could not, news opinions have diversified to capture niche segments of different audiences. Even mainstream news sources are popularly known to "lean" one direction or another, and specific personalities often acquire larger-than-life status among fans of these networks.

Our goal is to examine political polarization outside the realm of politics itself, but rather inside the modern news media. In order to accomplish this, we have scraped transcripts from shows hosted by two especially prominent news personalities, Rachel Maddow and Sean Hannity, as well as the less sensational Fox News Sunday show to act as a control.

These data sources can be found at:

    The Rachel Maddow Show: http://www.msnbc.com/transcripts/rachel-maddow-show
    
    Hannity: http://www.foxnews.com/on-air/hannity/transcripts
    
    Fox News Sunday: http://www.foxnews.com/on-air/fox-news-sunday-chris-wallace/transcripts

# Scraping the Data

Scraping data from the web presents unique challenges for data analysis, especially when that data is not inherently tabular or reguarly-structured at its source. While these news networks have loose standards that they tend to follow as they post these transcripts, there is still significant variance in ways that the text is written. This leads to complicated entity resolution problems, such as whether "RACHEL MADDOW, MSNBC HOST" and "RACHEL MADDOW, MSNBC ANCHOR" should be classified as the same speaker, whether "(BEGIN VIDEO CLIP)" and "(BEGIN VIDEO TAPE)" have the same meaning, and one-off problems such as what should be done with a transcript line for "(AUDIENCE APPLAUDS)."

The full scripts used to scrape the data can be found at: https://github.com/zacrobertson/EDAV_Project/blob/master/Parser_Functions.ipynb.

To briefly summarize the processes in that notebook, first we used the python library beautifulsoup to read the respective transcript URLs for each show, targeting shows since the beginning of 2017. Each episode of these shows was then fed to a series of parser functions which extract the text from the transcript web page, separate the speaker from their text, and then assign a series of attributes based on the text: number of words in the text, number of letters in the text, segments within each show (occurring between commercials, for the two Fox News shows only), whether the speaker was the show host or a guest, and whether the text was spoken by someone on the show or aired as part of a tape (again for the Fox News Shows only).

Because the sample size of show transcripts is rather large, but data quality of scraped text can be questionable, we opted for maximizing data quality by rejecting instances of each show that did not conform to some minimum standards. For example, some Rachel Maddow Show transcript lines have no listed speaker, and it may be obvious to a reader that this means Rachel Maddow herself is speaking, but it is substantially less obvious to the automatic scraper. Every instance of Fox News Sunday met the given standards, but not every episode of Hannity or The Rachel Maddow Show did-- a visualization of which shows are available will be displayed later in the notebook.
